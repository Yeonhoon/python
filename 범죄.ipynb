```python
crime = pd.read_csv('data/crime_seoul.csv', thousands=',', encoding = 'utf-8')
crime.dropna(inplace=True)
crime['']
```

```python
crime.tail()
crime.columns
```

```python
crime=crime.assign(강간검거율= lambda x: crime['강간강제 검거']/crime['강간강제발생']*100)
crime=crime.assign(강도검거율 = lambda x: crime['강도 검거']/crime['강도 발생']*100,
            살인검거율 = lambda x: crime['살인 검거']/crime['살인 발생']*100,
            절도검거율 = lambda x: crime['절도 검거']/crime['절도 발생']*100,
            폭력검거율 = lambda x: crime['폭력 검거']/crime['폭력 발생']*100)


```

```python
con_list = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']

for column in con_list:
    crime.loc[crime[column]>100, column] = 100

crime.sort_values(by='관서명', ascending=True, inplace=True)
crime.set_index('관서명',inplace=True)
```

```python
crime.rename(columns = {'강간강제발생': '강간',
                       '강도 발생': '강도',
                       '살인 발생': '살인',
                        '폭력 발생': '폭력',
                        '절도 발생': '절도'}, inplace=True)
```

```python
crime.tail()
```

```python
# 건수 정규화 하기
from sklearn import preprocessing

col = ['강간','강도','살인','절도','폭력']

x = crime[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_norm = pd.DataFrame(x_scaled, columns = col, index = crime.index)
col2  = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_norm[col2] = crime[col2]
```

```python
cctv_result = pd.read_csv('data/cctv_result.csv',encoding='utf-8')
cctv_result.set_index('구별', inplace=True)
```

```python
crime_norm[['인구수','cctv']]= cctv_result[['총 인구수','소계']]
```

```python
col = ['강간','강도','살인','절도','폭력']
crime_norm['범죄'] = np.sum(crime_norm[col], axis=1)
col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_norm['검거'] = np.sum(crime_norm[col2], axis=1)
```
