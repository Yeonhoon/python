```python
data_result=data_result.assign(cctv비율 = data_result['소계']/data_result['총 인구수']*100)
data_result['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['총 인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['총 인구수'], data_result['소계'],1) # polyfit: 직선 만들기
fp1
```

```python
f1 = np.poly1d(fp1) #poly1d: y축
fx = np.linspace(100000, 700000, 100) #linspace:x축
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['총 인구수'], data_result['소계'], s=50)
plt.xlabel('총 인구수')
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차']= np.abs(data_result['소계']) -f1(data_result['총 인구수'])
df_sort = data_result.sort_values(by="오차", ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(10,14))
plt.scatter(data_result['총 인구수'], data_result['소계'], c= data_result['오차'],s=50)
plt.xlabel('총 인구수')
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['총 인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
plt.colorbar()    
plt.ylabel('CCTV')
plt.grid()
plt.show()
```
